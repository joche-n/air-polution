```python
# all 0s are replaced with NaN, and then visualize the new missing values using a missingno bar plot

import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt

# Step 1: Create a copy of your original DataFrame
train_df_with_nan_zeros = train_df.copy()

# Step 2: Replace all 0s with NaN
train_df_with_nan_zeros = train_df_with_nan_zeros.replace(0, np.nan)

# Step 3: Visualize missing values with missingno
msno.bar(train_df_with_nan_zeros)
plt.title("Missing Values (Original NaNs + Zeros as NaNs)")
plt.show()
```

```python
# Count missing values in each column
missing_counts = train_df_with_nan_zeros.isna().sum()

# Keep only columns that have at least one missing value
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)

# Convert to DataFrame for nicer formatting
missing_df = missing_counts.to_frame(name='Missing Count')
missing_df['% Missing'] = (missing_df['Missing Count'] / len(train_df_with_nan_zeros) * 100).round(2)

# Display the full table
display(missing_df)
```

```python
# Count missing values
missing_counts = train_df_with_nan_zeros.isna().sum()

# Keep only columns with NO missing values
complete_counts = missing_counts[missing_counts == 0].sort_index()

# Format as a nice DataFrame
complete_df = complete_counts.to_frame(name='Missing Count')
complete_df['% Missing'] = 0.00  # Explicitly show zero missing percentage

# Display the result
display(complete_df)
```

```python
# DROP all angles except from AER_AI
# Step 1: Identify columns that contain "_angle" but NOT "L3_AER_AI"
cols_to_drop = [
    col for col in train_df_with_nan_zeros.columns 
    if "_angle" in col and "L3_AER_AI" not in col
]

# Step 2: Drop those columns and assign to a new DataFrame
train_cleaned = train_df_with_nan_zeros.drop(columns=cols_to_drop)

# Step 3: Print dropped columns and new shape
print("Dropped columns:")
print(cols_to_drop)
print("\nOriginal shape:", train_df_with_nan_zeros.shape)
print("Cleaned shape:", train_cleaned.shape)
```

```python
train_cleaned.head()
```

```python
# Count missing values
missing_counts = train_cleaned.isna().sum()

# Filter to show only columns with missing values
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)

# Format as a DataFrame
missing_df = missing_counts.to_frame(name='Missing Count')
missing_df['% Missing'] = (missing_df['Missing Count'] / len(train_cleaned) * 100).round(2)

# Display the result
display(missing_df)
```

```python
# List of columns to drop (those with >18% missing)
columns_to_drop_18plus = [
    'L3_CH4_aerosol_optical_depth',
    'L3_CH4_aerosol_height',
    'L3_CH4_CH4_column_volume_mixing_ratio_dry_air',
    'L3_NO2_tropospheric_NO2_column_number_density',
    'L3_HCHO_cloud_fraction',
    'L3_SO2_cloud_fraction',
    'L3_HCHO_HCHO_slant_column_number_density',
    'L3_HCHO_tropospheric_HCHO_column_number_density_amf',
    'L3_HCHO_tropospheric_HCHO_column_number_density',
    'L3_SO2_absorbing_aerosol_index',
    'L3_SO2_SO2_column_number_density_amf',
    'L3_SO2_SO2_slant_column_number_density',
    'L3_SO2_SO2_column_number_density',
    'L3_CO_cloud_height',
    'L3_CO_sensor_altitude',
    'L3_CO_CO_column_number_density',
    'L3_CO_H2O_column_number_density'
]

# Drop from train_cleaned and create new DataFrame
train_reduced = train_cleaned.drop(columns=columns_to_drop_18plus)

# Print result
print("Dropped columns with >18% missing values:")
print(columns_to_drop_18plus)
print("\nNew shape:", train_reduced.shape)
```
